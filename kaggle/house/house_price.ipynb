## scikit-learn 선형회귀 모델(without 전처리)

```python
X = df[['sqft_living','grade','sqft_above','sqft_living15']]
X.head()
```

```python
Y = df['price']
Y.head()
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X,Y)
```

```python
pre = lr.predict(X)
plt.scatter(pre,Y, s=10)
plt.xlabel("Predicted price")
plt.ylabel("Actual price")

plt.show()
```

전처리를 하지 않으니 실제값과 많이 차이 나는것 같다. 엉엉

### Normalization

```python
X.head()
```

```python
X_norm = X

for a in X.columns:
    f_max = X[a].max()
    X_norm[a] = X_norm[a].apply(lambda x: x/f_max)
X_norm.head()
```

```python
lr_norm = LinearRegression()
lr_norm.fit(X_norm,Y)

pre_norm = lr.predict(X_norm)
plt.scatter(pre_norm,Y, s=10)
plt.xlabel("Predicted price")
plt.ylabel("Actual price")

plt.show()
```
