Pendulum 문제

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
N =1000
```

```python
y = np.zeros([4])
```

```python
L_o = 1
L = 1
v_o =0
theta_o = 0.3
omega_o = 0
```

```python
y[0] = L
y[1] = v_o
y[2] = theta_o
y[3] = omega_o
```

```python
time = np.linspace(0,25, N)
```

```python
def spring_pendulum(y,time):
    k = 3.5
    m = 0.2
    gravity = 10
    
    L_o = 1
    L = 1
    v_o =0
    theta_o = 0.3
    omega_o = 0
    
    g0 = y[1]
    g1 = (L_o+y[0])*y[3]*y[3]-k/m*y[0]+gravity*np.cos(y[2])
    g2 = y[3]
    g3 = -(gravity*np.sin(y[2])+2*y[1]*y[3])/(L_o+y[0])
    
    return np.array([g0,g1,g2,g3])
```

```python
answer = odeint( spring_pendulum , y , time )
```

```python
xdata = (L_o + answer[:,0])*np.sin(answer[:,2])
ydata = -(L_o + answer[:,0])*np.cos(answer[:,2])
```

```python
plt.plot(xdata,ydata, 'r-')
plt.xlabel("Horizontal position")
plt.ylabel("Vertical position")
plt.show()
```
