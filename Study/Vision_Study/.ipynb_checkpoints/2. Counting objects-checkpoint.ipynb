```python
import cv2
import imutils
import matplotlib.pyplot as plt
import time

%matplotlib inline
```

```python
image = cv2.imread('./image/tetris_blocks.png')

plt.imshow(image)
```

## Converting an image to grayscale

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap='gray')
```

## Edge detection

Canny algorithm을 사용:
    - img : gray 이미지
    - minVal : 최소 threshold
    - maxval : 최대 threshold
    - aperutre_size : The Sobel 커널 사이즈, 기본은 3

```python
edged = cv2.Canny(gray, 30, 150)

plt.imshow(edged, cmap='gray')
```

### Thresholding
Thresholding을 하면 더 가볍고 어두운 영상 영역과 윤곽선을 제거할 수 있다.

임계값(255) 보다 크면 흰색(255), 작으면 검은색으로 처리하자




```python
thresh = cv2.threshold(gray, 225, 255, cv2.THRESH_BINARY_INV)[1]

plt.imshow(thresh, cmap='gray')
```

## Detecting and drwaing contours


find contours (i.e., outlines) of the foreground objects in the thresholded image.

```python
cnts = cv2.findContours(thresh.copy(), cv2.RETR_EXTERNAL,
                        cv2.CHAIN_APPROX_SIMPLE)
print(cnts)
```

cv2.findContours를 이용하여 윤곽선을 찾는다: 우리는 thresh.copy의 white pixel을 찾는다.

```python
cnts = imutils.grab_contours(cnts)
print(cnts)
```

imutils.grab_coutours로 꼭짓점만 모아준것 같기도 하고...


```python
output = image.copy() # image를 복사해옴
 
# loop over the contours
for c in cnts:
    # draw each contour on the output image with a 3px thick purple
    # outline, then display the output contours one at a time
    cv2.drawContours(output, [c], -1, (240, 0, 159), 3)
    print([c])
    plt.imshow(output)
    plt.show()
    time.sleep(1)
    
```

image를 복사해와서 output이라고 했고, output에 윤곽선을 그렸다.

cv2.drawContours는 윤곽선을 그려준다

```python
text = "I found {} objects!".format(len(cnts))
cv2.putText(output, text, (10, 25),  cv2.FONT_HERSHEY_SIMPLEX, 0.7,
            (240, 0, 159), 2)

plt.imshow(output)
```

총 찾은 물체 갯수를 추가했다.

## Erosions and dilations

Erosions and dilations은 노이즈를 줄이기 위해 사용한다.

object의 크기를 줄이기 위해, 반복적으로 픽셀을 erode한다.

```python
mask = thresh.copy()
mask = cv2.erode(mask, None, iterations=5)

plt.imshow(mask, cmap='gray')
```

thresh를 복사해와 mask라고 이름붙였다.

cv2.erode로 5번 반복하여 윤곽선의 크기를 줄였다.


```python
mask = thresh.copy()
mask = cv2.dilate(mask, None, iterations=5)

plt.imshow(mask, cmap='gray')
```

비슷하게 넓힐 수도 있다.

### Masking and bitwise operations

마스크를 사용하면 관심이없는 이미지 영역을 "마스크 아웃"할 수 있습니다. 

우리가 신경 쓰지 않는 이미지 영역을 숨기므로 "마스크"라고합니다.

a typical operation we may want to apply is to take our mask and apply a bitwise "AND" to our input image, keeping only the masked regions


우리가 적용할 수 있는 전형적인 수술은 마스크를 쓰고 우리의 입력 이미지에 약간 "AND"를 적용하여 마스크된 부분만 유지하는 것이다.

```python
mask = thresh.copy()
output = cv2.bitwise_and(image, image, mask=mask)

plt.imshow(output)
```
