```python
import cv2
import imutils
import matplotlib.pyplot as plt
%matplotlib inline
```

## 이미지 불러오기

```python
image = cv2.imread('./image/다운로드.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

(h,w,d) = image.shape
print("width = {}, height = {}, depth = {}".format(w,h,d))
plt.imshow(image)
```

```python
# access the RGB pixel located at x=50, y=100, keepind in mind that
# OpenCV stores images in BGR order rather than RGB
(R, G, B) = image[100, 50]
print("R={}, G={}, B={}".format(R, G, B))
```

## Roi

```python
roi = image[80:180,37:137] #100px X 100 px
plt.imshow(roi)
```

## Resizing

```python
r = 300.0 / w
dim = (300, int(h*r))

resized = cv2.resize(image,dim)
plt.imshow(resized)
```

## Resizing by imutils

```python
resized = imutils.resize(image, width = 500)

plt.imshow(resized)
```

## Rotating image
이미지를 45도 돌려보자
### 1. opencv

```python
center = (w // 2, h//2)
print(center)
```

```python
#center를 중심으로 45도 시계방향 회전시키는 행렬 만듦
M = cv2.getRotationMatrix2D(center, -45, 1.0)
print(M)
```

```python
rotated = cv2.warpAffine(image, M, (w,h))

plt.imshow(rotated)
```

### 2. imutils

```python
rotated = imutils.rotate(image, -45)

plt.imshow(rotated)
```

### 3. imtuils bound

```python
rotated = imutils.rotate_bound(image, 45)

plt.imshow(rotated)
```

## smoothing

```python
blurred = cv2.GaussianBlur(image,(7,7),0)
# 커널이 클 수록 더 스무딩함...

plt.imshow(blurred)
```

## Draw on an image
### Rectangel

```python
output = image.copy()
plt.imshow(output)
```

```python
cv2.rectangle(output,(75,25),(190,125),(0,0,255),thickness=2) 
#(60,60) : stating pixel , (160,160) : ending pixel, (0,0,255): color, 2: thickness

plt.imshow(output)
```

### 2. Circle

```python
output = image.copy()
cv2.circle(img = output,center = (60,60),radius=20,color=(255,0,0), thickness=-1) 
#thickness nagative -> solid/filled

plt.imshow(output)
```

### 3. Line

```python
output = image.copy()
cv2.line(img=output, pt1=(60,20), pt2=(160,160),color=(0,0,255),thickness=5)

plt.imshow(output)
```

### 4. Text

```python
output = image.copy()
text = "Opencv Studying"
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img=output, text=text, org=(10,25),fontFace=font, fontScale=0.7, color=(0,0,255),thickness=2)

plt.imshow(output)
```
